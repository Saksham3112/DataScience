```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=pd.read_csv('Diabetes_Xtrain.csv')
y=pd.read_csv('Diabetes_Ytrain.csv')
test=pd.read_csv('Diabetes_Xtest.csv')
```

```python
x=x.values
y=y.values
test=test.values
## BY DROPPING WE GET 77% at K=10
## MAXIMUM WITHOUT DROPPING AT k=17
x=x[:,(0,1,4,5,6,7)]
test=test[:,(0,1,4,5,6,7)]
```

```python
plt.style.use('seaborn')
for i in range(len(x[0])-1):
    plt.scatter(x=x[:,i],y=y,marker='*')
    plt.title(str(pd.DataFrame(x[:,i]).corrwith(pd.DataFrame(y))))
    plt.show()
```

```python
def predict_ (p,k):
    dis=[]
    for i in range(len(x)):
        cdist=0
        for j in range(1,len(p)):
            cdist+=(p[j]-x[i][j])**2
        dis.append([cdist**0.5,i])
    dis=sorted(dis,key=lambda x:x[0])
    count=0
    for i in range(k):
        if(y[dis[i][1]]==1):
            count+=1
    if(count>k/2):
        return 1
    else:
        return 0
```

```python
a=open('Output.csv',mode='w')
k=9
a.write('Outcome\n')
for i in range(len(test)):
    a.write(str(predict_(test[i],k)))
    a.write('\n')
a.close()
```
